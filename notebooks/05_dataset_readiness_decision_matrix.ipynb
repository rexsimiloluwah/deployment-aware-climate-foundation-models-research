## 2. Practical Readiness Scores

```python
scores = pd.DataFrame([
    {'dataset': 'ftw_africa', 'role': 'main benchmark', 'readiness_score': 8, 'n_labeled': 1137},
    {'dataset': 'sen1floods11_ghana', 'role': 'secondary pilot', 'readiness_score': 5, 'n_labeled': 53},
])
display(scores)

fig, ax = plt.subplots(figsize=(7, 4))
ax.bar(scores['dataset'], scores['readiness_score'])
ax.set_ylim(0, 10)
ax.set_title('Dataset Readiness for LoRA vs Linear Probe Experiments')
ax.set_ylabel('Practical Readiness Score / 10')
ax.set_xlabel('Dataset')
ax.grid(axis='y', alpha=0.25)
for idx, row in scores.iterrows():
    ax.text(idx, row['readiness_score'] + 0.2, f"n={row['n_labeled']}", ha='center')
fig.tight_layout()
```

## Go / No-Go Rule

Commit a dataset to supervised model experiments only when these are true:

- Data loader produces stable train/val/test batches.
- Labels pass visual or distribution sanity checks.
- 10%, 25%, and 100% label budgets remain meaningful.
- Metrics are defensible for the task.
- The African climate-resilience story is clear enough for the lightning talk.

Current read: `ftw_africa` is the main supervised benchmark. `sen1floods11_ghana` is a small but useful disaster-response pilot.
